```python
# cell1
# Import the required libraries
import pandas as pd
import numpy as np
import yfinance as yf
import gc
import joblib

# from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, roc_auc_score
```

```python
# cell2
# Download data for a bunch of stocks
tickers = [
    # Tech
    "AAPL", "MSFT", "GOOGL", "AMZN", "META", "NVDA", "TSLA", "AMD", "INTC", "IBM",
    "ORCL", "CRM", "ADBE", "CSCO", "QCOM", "AVGO", "TXN", "SHOP", "PYPL", "NFLX",

    # Finance
    "JPM", "BAC", "WFC", "C", "GS", "MS", "AXP", "BLK", "SCHW", "COF",

    # Healthcare
    "JNJ", "PFE", "MRK", "ABBV", "TMO", "ABT", "LLY", "BMY", "GILD", "CVS",

    # Consumer (Retail / Staples)
    "WMT", "TGT", "HD", "LOW", "COST", "MCD", "SBUX", "KO", "PEP", "PG",
    "CL", "KHC", "MDLZ", "UL", "DEO", "NKE", "LULU", "DIS", "CMG", "YUM",

    # Energy
    "XOM", "CVX", "COP", "EOG", "SLB", "BP", "SHEL", "HAL", "PSX", "OXY",

    # Industrials
    "GE", "BA", "CAT", "DE", "UNP", "UPS", "FDX", "MMM", "LMT", "RTX",

    # Utilities
    "NEE", "DUK", "SO", "AEP", "EXC", "D", "SRE", "XEL", "PEG", "WEC",

    # Telecom
    "VZ", "T", "TMUS", "CHTR", "CMCSA",

    # Real Estate / Misc
    "PLD", "AMT", "SPG", "EQIX", "CCI",
    "BRK-B",  # skip BRK-A
    "WBD", "ETSY", "ROKU",
]

frames = []  # we'll store each ticker's DataFrame here

for t in tickers:
    print(f"Downloading {t}...")
    df = yf.download(t, period="10y")

    if df.empty:
        print(f"  -> no data for {t}, skipping.")
        continue

    #  NEW: flatten MultiIndex columns (('Close','AAPL')  'Close')
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    df["Ticker"] = t        # add ticker column
    frames.append(df)       # save this ticker's data

# stack all tickers vertically into one long table
data = pd.concat(frames, axis=0)
data = data.reset_index()   # Date becomes a normal column
data.head()
```

```python
# cell4
# ---------- Feature Engineering (lighter version) ----------

# RSI helper
def compute_rsi(close, window=14):
    delta = close.diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)

    avg_gain = gain.rolling(window).mean()
    avg_loss = loss.rolling(window).mean()

    rs = avg_gain / avg_loss
    rsi = 100 - (100 / (1 + rs))
    return rsi

# Download SP500 (SPY-ish) for market context
spy = yf.download("^GSPC", period="10y")
spy = spy.sort_index()
spy_ret_5d = spy["Close"].pct_change(5)
spy_ret_10d = spy["Close"].pct_change(10)

frames_feat = []

for df_t in frames:
    ticker = df_t["Ticker"].iloc[0]

    # sort by date
    df_t = df_t.sort_index().copy()

    # ---------- Base price features ----------
    df_t["return_1d"]  = df_t["Close"].pct_change()
    df_t["return_5d"]  = df_t["Close"].pct_change(5)
    df_t["return_10d"] = df_t["Close"].pct_change(10)

    daily_ret = df_t["Close"].pct_change()
    df_t["vol_5d"]     = daily_ret.rolling(5).std()

    df_t["ma_5"]       = df_t["Close"].rolling(5).mean()
    df_t["ma_20"]      = df_t["Close"].rolling(20).mean()
    df_t["ma_ratio"]   = df_t["ma_5"] / df_t["ma_20"]

    # ---------- RSI (14-day) ----------
    df_t["rsi_14"] = compute_rsi(df_t["Close"], window=14)

    # ---------- Volume features ----------
    df_t["vol_ma_10"] = df_t["Volume"].rolling(10).mean()
    df_t["vol_ratio"] = df_t["Volume"] / df_t["vol_ma_10"]

    # ---------- Market context: SP500 relative strength ----------
    df_t["spy_ret_5d"] = spy_ret_5d.reindex(df_t.index)
    df_t["spy_ret_10d"] = spy_ret_10d.reindex(df_t.index)

    df_t["rel_strength_5d"]  = df_t["return_5d"]  - df_t["spy_ret_5d"]
    df_t["rel_strength_10d"] = df_t["return_10d"] - df_t["spy_ret_10d"]

    # ---------- Targets: 1-day and 5-day future direction ----------
    df_t["close_1d_ahead"] = df_t["Close"].shift(-1)
    df_t["ret_1d_ahead"]   = (df_t["close_1d_ahead"] - df_t["Close"]) / df_t["Close"]
    df_t["up_1d"]          = (df_t["ret_1d_ahead"] > 0).astype(int)

    df_t["close_5d_ahead"] = df_t["Close"].shift(-5)
    df_t["ret_5d_ahead"]   = (df_t["close_5d_ahead"] - df_t["Close"]) / df_t["Close"]
    df_t["up_5d"]          = (df_t["ret_5d_ahead"] > 0).astype(int)

    df_t["Ticker"] = ticker
    frames_feat.append(df_t)

# Stack all tickers back together
data = pd.concat(frames_feat, axis=0)

# Reset index so Date becomes a normal column again
data = data.reset_index()

# Normalize Date column name
if "index" in data.columns and "Date" not in data.columns:
    data = data.rename(columns={"index": "Date"})

# Ensure Date is datetime
data["Date"] = pd.to_datetime(data["Date"])

# Drop rows with NaNs from rolling/targets/spy alignment
data = data.dropna()

# Optional: keep only recent years
data = data[data["Date"] >= "2018-01-01"]

# Define feature set (lighter)
feature_cols = [
    "return_1d", "return_5d", "return_10d",
    "vol_5d",
    "ma_5", "ma_20", "ma_ratio",
    "rsi_14",
    "vol_ma_10", "vol_ratio",
    "spy_ret_5d", "spy_ret_10d",
    "rel_strength_5d", "rel_strength_10d",
]

X = data[feature_cols]
y_1d = data["up_1d"]
y_5d = data["up_5d"]

data.head()
```

```python
# cell6
import os
import warnings
from sklearn.exceptions import UndefinedMetricWarning
from sklearn.model_selection import TimeSeriesSplit

# Ignore AUC warnings when test set has only 0s or only 1s
warnings.filterwarnings("ignore", category=UndefinedMetricWarning)

# Make sure output folders exist
os.makedirs("models_1d", exist_ok=True)
os.makedirs("models_5d", exist_ok=True)

ticker_scores_1d = []
ticker_scores_5d = []

tickers_unique = sorted(data["Ticker"].unique())

tscv = TimeSeriesSplit(n_splits=2)   # keep lightweight

# OPTIONAL: while testing, limit to first N tickers
# tickers_unique = tickers_unique[:10]

for ticker in tickers_unique:
    print(f"\n===== {ticker} =====")

    df_t = data[data["Ticker"] == ticker].sort_values("Date").copy()
    if len(df_t) < 300:
        print("  Skipped (not enough data)")
        continue

    X_t  = df_t[feature_cols]
    y1_t = df_t["up_1d"]
    y5_t = df_t["up_5d"]

    # Use last 20% as a simple test set
    split_idx = int(len(df_t) * 0.8)
    X_test_1d = X_t.iloc[split_idx:]
    y_test_1d = y1_t.iloc[split_idx:]
    y_test_5d = y5_t.iloc[split_idx:]

    # =====================
    # 1-DAY MODEL
    # =====================
    xgb_1d = XGBClassifier(
        n_estimators=120,
        max_depth=3,
        learning_rate=0.07,
        subsample=0.8,
        colsample_bytree=0.8,
        n_jobs=-1,
        eval_metric="logloss",
        random_state=42,
    )
    xgb_1d.fit(X_t, y1_t)
    joblib.dump(xgb_1d, f"models_1d/{ticker}.pkl")

    preds1 = xgb_1d.predict(X_test_1d)
    prob1  = xgb_1d.predict_proba(X_test_1d)[:, 1]

    acc_1d = accuracy_score(y_test_1d, preds1)
    try:
        auc_1d = roc_auc_score(y_test_1d, prob1)
    except ValueError:
        auc_1d = np.nan

    print(f"  1-day Acc={acc_1d:.3f}, AUC={auc_1d:.3f}")
    ticker_scores_1d.append((ticker, acc_1d, auc_1d))

    # =====================
    # 5-DAY MODEL
    # =====================
    xgb_5d = XGBClassifier(
        n_estimators=120,
        max_depth=3,
        learning_rate=0.07,
        subsample=0.8,
        colsample_bytree=0.8,
        n_jobs=-1,
        eval_metric="logloss",
        random_state=42,
    )
    xgb_5d.fit(X_t, y5_t)
    joblib.dump(xgb_5d, f"models_5d/{ticker}.pkl")

    preds5 = xgb_5d.predict(X_test_1d)
    prob5  = xgb_5d.predict_proba(X_test_1d)[:, 1]

    acc_5d = accuracy_score(y_test_5d, preds5)
    try:
        auc_5d = roc_auc_score(y_test_5d, prob5)
    except ValueError:
        auc_5d = np.nan

    print(f"  5-day Acc={acc_5d:.3f}, AUC={auc_5d:.3f}")
    ticker_scores_5d.append((ticker, acc_5d, auc_5d))

    # Cleanup to save RAM
    del df_t, X_t, y1_t, y5_t, X_test_1d, y_test_1d, y_test_5d
    del xgb_1d, xgb_5d, preds1, preds5, prob1, prob5
    gc.collect()
```

```python
# cell10
import joblib
import yfinance as yf

def prepare_live_features(ticker):
    """
    Build the latest feature row for a given ticker,
    using the SAME logic as the training features (Cell 4).
    """
    # 1) Download recent history for this ticker
    df = yf.download(ticker, period="200d")
    df = df.sort_index()

    #  NEW: flatten MultiIndex columns if needed
    # (e.g., ('Close','AAPL') -> 'Close')
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = df.columns.get_level_values(0)

    # --- price-based features (match Cell 4) ---
    df["return_1d"]  = df["Close"].pct_change()
    df["return_5d"]  = df["Close"].pct_change(5)
    df["return_10d"] = df["Close"].pct_change(10)

    daily = df["Close"].pct_change()
    df["vol_5d"] = daily.rolling(5).std()
    df["ma_5"]   = df["Close"].rolling(5).mean()
    df["ma_20"]  = df["Close"].rolling(20).mean()
    df["ma_ratio"] = df["ma_5"] / df["ma_20"]

    # --- RSI(14) ---
    delta = df["Close"].diff()
    gain = delta.clip(lower=0)
    loss = -delta.clip(upper=0)
    avg_gain = gain.rolling(14).mean()
    avg_loss = loss.rolling(14).mean()
    rs = avg_gain / avg_loss
    df["rsi_14"] = 100 - (100 / (1 + rs))

    # --- volume features ---
    # After flattening, df["Volume"] should now be a Series
    df["vol_ma_10"] = df["Volume"].rolling(10).mean()
    df["vol_ratio"] = df["Volume"] / df["vol_ma_10"]

    # --- SP500 context (same as Cell 4) ---
    spy = yf.download("^GSPC", period="200d").sort_index()

    # flatten spy cols too, just in case
    if isinstance(spy.columns, pd.MultiIndex):
        spy.columns = spy.columns.get_level_values(0)

    spy5  = spy["Close"].pct_change(5)
    spy10 = spy["Close"].pct_change(10)

    df["spy_ret_5d"]  = spy5.reindex(df.index)
    df["spy_ret_10d"] = spy10.reindex(df.index)
    df["rel_strength_5d"]  = df["return_5d"]  - df["spy_ret_5d"]
    df["rel_strength_10d"] = df["return_10d"] - df["spy_ret_10d"]

    # Drop rows with NaNs in feature columns
    df = df.dropna(subset=feature_cols)

    # Take the most recent row of features
    latest_feats = df[feature_cols].iloc[-1:]
    latest_date  = df.index[-1]
    return latest_feats, latest_date


def predict_live(ticker):
    """
    Use the saved XGBoost models for this ticker
    to get 1-day and 5-day up probabilities.
    """
    X_live, live_date = prepare_live_features(ticker)

    model_1d = joblib.load(f"models_1d/{ticker}.pkl")
    model_5d = joblib.load(f"models_5d/{ticker}.pkl")

    p1 = model_1d.predict_proba(X_live)[0, 1]
    p5 = model_5d.predict_proba(X_live)[0, 1]

    print(f"\n=== Live Prediction for {ticker} (as of {live_date.date()}) ===")
    print(f"Prob UP Tomorrow:  {p1:.3f}")
    print(f"Prob UP in 5 Days: {p5:.3f}")

    return p1, p5
```

```python
# cell8
predict_live("AAPL")
predict_live("MSFT")
predict_live("NVDA")
```
